## 생물학적 뉴런에서 인공 뉴런까지

#### 뉴런을 사용한 논리연산

하나 이상의 이진(on/off)입력고 이진 출력 하나를 가진다. 인공뉴런은 단순히 입력이 일정 개수만큼 활성화 되었을 때 출력을 내보낸다. 인공 뉴런의 네트워크를 만들어 어떤 논리 명제도 계산할 수 있다는 것을 증명했다. 

인공신경망.jpg

첫번째 네트워크는 항등 함수다. 뉴런A가 활성화 되면 뉴런 C도 활성화 된다. 하지만 뉴런A가 꺼지면 C도 꺼진다.

두번째 네트워크는 논리곱 연산을 수행한다. 뉴런 A,B 모두가 활성화 될때만 뉴런 C가 활성화된다.(입력신호 하나로는 C를 활성화 못한다.)

세번째 네트워크는 논리 합 연산을 수행한다. A,B 중 하나가 활성화 되면 C가 활성화된다.

어떤 입력이 뉴런의 활성화를 억제할 수 있다고 가정하면 네번째 네트워크는 조금 더 복잡한 논리 명제를 계산할 수 있다. A가 활성화 되고 B가 비활성화 될때 C가 활성화 된다. 만약 뉴런 A 가 항상 활성화 되어 있다면 이 네트워크는 논리 부정연산이 된다. B가 비활성화될 때 C가 활성화 되고, 또는 정반대로 B가 활성화될때 C가 비활성화된다.

### 퍼셉트론

가장 간단한 인공 신경망 구조 중 하나다. 입 출력이 어떤 숫자이고, 각각의 입력 연결은 가중치와 연관되어있다. 입력의 가중치 합을 계산한뒤 계산된 합에 계단 함수를 적용하여 결과를 출력한다.

퍼셉트론.jpg

퍼셉트론에서 가장 널리 사용도는 계단 함수는 헤비사이드 계단 함수다. 부호함수를 대신 사용할 때가 있다.

계단함수.jpg

입력층은 모두 입력 뉴런으로 구성된다. 보통 편향 특성이 더해진다. 편향특성은 항상 1을 출력하는 특별한 종류의 뉴런인 편향 뉴런으로 표현된다.

퍼셉트론구조.jpg

출력계산.jpg

X는 입력 특성의 행렬을 나타낸다. 행은 샘플 열은 특성이다.

가중치 행렬 W는 편향뉴런을 제외한 모든 연결 가중치를 포함한다. 행은 입력뉴런 열은 출력층에 있는 인공뉴런이다.

편향 벡터 b는 편향 뉴런과 인공 뉴런 사이의 모든 연결 가중치를 포함한다. 인공뉴런마다 하나의 편향값이 있다.

활성화 함수라 한다.

네트워크가 예측을 할 때 만드는 오차를 반영하도록 조금 변형된 규칙을 사용하여 훈련된다. 오차가 감소되도록 연결을 강화한다. 잘못된 예측을 하는 모든 출력 뉴런에 대해 올바른 예측을 만들 수 있도록 입력에 연결된 가중치를 강화시킨다.

학습규칭.jpg

### 다층 퍼셉트론과 역전파

다층 퍼셉트론은 입력층 하나와 은닉층이라 불리는 하나이상의 TLU층과 마지막 출력층으로 구성된다. 입력층과 가까운 층을 보통 하위층 이라 부르고 출력층과 가까운 층을 상위층 이라한다. 출력증을 제외하고 모든 층은 편향 뉴런을 포함하며 다음 층과 완전히 연결되있다.

입력층.jpg

은닉층을 여러개 쌓아 올린 심층 신경망(DNN)이라 한다.

다층 퍼셉트론을 훈련할 획기적인 방법은 역전파 이다. 효율적인 기법으로 그레이디언트를 자동계산하는 경사 하강법이다. 네트워크를 두번(정방향 역방향)통과 하는것만으로 모든 모델 파라미터에 대한 오차의 그레이디언트를 계산할수있다.

기여도는 가중치의 미분값이다

하나의 미니배치씩 지행하여 전체 훈련 세트를 처리한다. 각 반복을 에포크(epoch)라고 한다.

각 미니배치는 입력층으로 전달되어 마지막 출력층의 출력을 계산하는데 이를 정방향 계산이라 한다.

각 출력 연결이 이 오차에 기여하는 정도를 계산한다. 이를 연쇄법칙(chain rule)을 적용하면 이 단계를 바르고 정확하게 수행할 수 있다.

가중치수정(경사하강법)

연쇄 법칙을 사용하여 이전층의 연결 가중치가 이 오차의 기여 정도에 얼마나 기여했는지 측정한다. 이런 역방향 단계는 오차 그레이디언트를 거꾸로 전파함으로써 효율적으로 네트워크에 있는 모든 연결 가중치에 대한 오차 그레이디언트를 측정한다.

활성화 함수.jpg

비선형함수를 추가하지 않으면 아무리 많은 층을 쌓아도 하나의 층과 동일해진다. 반대로 비선형 활성화 함수가 있는 충분히 큰 심층 신경망은 이론적으로 어떤 연속 함수도 근사할 수 있다.

### 회귀를 위한 다층 퍼셉트론

다층 퍼셉트론은 회귀 작업에 사용할 수 있다. 값 하나를 예측하는 데 출력 뉴런이 하나만 필요하다.

다변량 회귀에서는 출력 차원마다 출력 뉴런이 하나씩 필요하다.

일반적으로 회귀용 다층 퍼셉트론을 만들때 출력 뉴런에 활성화 함수를 사용하지 않고 어떤 범위의 값도 출력되도록 한다. 하지만 출력이 항상 양수여야 한다면 출력층에 ReLU함수를 사용할수있다. softplus함수도 사용할수잇다. softplus(z) = log(1+exp(z)). z 가 음수일 때 0에 가까워지고 z가 큰 양수일수록 z에 가까워진다.

어떤 범위 안의 값을 예측하고 싶다면 로지스틱 함수나 하이퍼볼릭 탄젠트함수를 사용하고 레이블의 스케일을 적절한 범위로 조정할수있다. 로지스틱 함수는 0에서 1 사이를 출력하고 하이퍼볼릭 탄젠트는 -1에서 1 사이를 출력한다.

훈련에서 사용하는 손실함수는 전형적으로 평균 제곱 오차다. 하지만 이상치가 많다면 평균 절댓값오차를 사용할 수 있다.

### 분류를 위한 다층 퍼셉트론

이진 분류 문제에서는 로지스틱 활성화 함수를 가진 하나의 출력 뉴런만 필요하다. 출력은 0~1사이의 실수다.

다중 레이블 이진분류문제를 쉽게 처리할수 있다. 로지스틱 활성화 함수를 가진 두 출력 뉴런이 필요하다. 첫번째 뉴런은 이메일이 스팸일 확률을 출력하고 두번째 뉴런은 긴급한 메일일 확률을 출력한다. 출력 확률값이 1일필요는 없다.

각 샘플이 3개 이상의 클래스중 한 클래스에만 속할숭 ㅣㅆ다면 클래스마다 하나의 출력뉴런이 필요하다. 출력층에는 softmax함수를 사용해야한다. 소프트맥스는 모든 예측확률을 0과 1사이에 두고 더했을때 1이 된다. 이를 다중분류라 한다.

#### 시퀀셜API를 사용하여 모델 만들기

```
model = keras.models.Sequential()
model.add(keras.layers.Flatten(input_shape=[28,28]))
model.add(keras.layers.Dense(300, activation='relu'))
model.add(keras.layers.Dense(100, activation='relu'))
model.add(keras.layers.Dense(10,activation='softmax'))
```



*   첫 번째 라인은 시퀀셜 모델을 만든다. 가장 간단한 케라스의 신경망 모델이다. 순서대로 연결된 층을 일렬로 쌓아서 구성한다.

*   첫번째 층을 만들고 모델에 추가한다. Flatten층은 입력 이미지를 1D배열로 변환한다. 즉 입력X를 받으면 X.reshape(-1,1)을 계산한다. 어떤 모델 파라미터도 가지지 않고 간단한 전처리를 수행할 뿐이다.
*   모델의 첫번째 층이니input_shape[]를 지정해야한다. keras.layers.InputLayer (?)층을 추가할수 있다.
*   뉴런 300개를 가진 은닉층을 추가한다. 이 층은 ReLU함수를 사용한다. Dense층마다 각자 가중치 행렬을 관리한다. 층의 뉴런과 입력 사이의 모든 연결 가중치가 포함되있다. 뉴런마다 하나씩 있는 편향도 벡터로 관리한다.


*   뉴런 100개를 가진 두번째 Dense은닉층을 추가한다. 역시 ReLU함수를 사용한다.


*   클래스마다 하나씩 뉴런 10개를 가진 Dense출력층을 추가한다. 소프트맥스 함수를 사용한다.

#### 모델 컴파일

모델을 만들고 compile()메서드를 호출하면 사용할 손실 함수와 옵티마이저를 지정해야 한다.

```
model.compile(loss = 'sparse_categorical_crossentropy', # 손실함수 크로스엔트로피 사용
              optimizer='sgd', #옵티마이저 확률적 경사하강법
              metrics=['accuracy']) # 분류기는 정확도 측정을 위해 사용
```

레이블이 정수 하나로 이루어져있다.(샘플마다 타깃 클래스 인덱스 하나가 있다.) 클래스가 배타적이므로 sparse_categorical_crossentropy를 사용한다.

옵티마이저에서 'sgd를 지정한다. 케라스가 앞서 소개한 역전파 알고리즘을 수행한다.

### 10.2.4 함수형 API를 사용해 복잡한 모델 만들기

순차적이지 않은 신경망의 한 예는 와이드&딥 신경망이다. 이 신경망 구조는 입력의 일부 또는 전체가 출력층에 바로 연결된다. 북잡한 패턴과 간단한 규칙을 모두 학습할수있다.

와이드.jpg

```
input_ = keras.layers.Input(shape=X_train.shape[1:])
hidden1 = keras.layers.Dense(30, activation ='relu')(input_)
hidden2 = keras.layers.Dense(30, activation ='relu')(hidden1)
concat = keras.layers.Concatenate()([input_, hidden2])
output = keras.layers.Dense(1)(concat)
model = keras.Model(inputs=[input_], outputs=[output])
```



*   Input객체를 만든다. shape와 dtype을 포함하여 모델의입력을 정의한다. 한 모델은 여러개의 입력을 가질 수 있다.

*   30개의 뉴련과 relu함수를 가진 Dense층을 만든다. 이 층은 입력과 함께 함수처럼 호출된다.

*   두 번째 은닉층을 만들고 함수처럼 호출한다.
*   Concatenate층을 만들고 함수처럼 호출하여 두 번째 은닉층의 출력과 입력을 연결한다. Concatenate층을 만들고 주어진 입력으로 바로 호출한다.


*   하나의 뉴런과 활성화 함수가 없는 출력층을 만들고 Concatenate층이 만든 결과를 사용해 호출한다.


*   마지막으로 사용할 입력과 출력을 지정하여 캐라스 Model을 만든다.



```
input_A = keras.layers.Input(shape=[5], name='wide_input')
input_B = keras.layers.Input(shape=[6], name='deep_input')
hidden1 = keras.layers.Dense(30, activation ='relu')(input_B)
hidden2 = keras.layers.Dense(30, activation ='relu')(hidden1)
concat = keras.layers.Concatenate()([input_A, hidden2])
output = keras.layers.Dense(1, name='output')(concat)
model = keras.Model(inputs=[input_A, input_B], outputs=[output])
```

여러개 입력.jpg

```
model.compile(loss='mse', optimizer=keras.optimizers.SGD(lr=1e-3))

X_train_A, X_train_B = X_train[:, :5], X_train[:, 2:]
X_valid_A, X_valid_B = X_valid[:, :5], X_valid[:, 2:]
X_test_A, X_test_B = X_test[:, :5], X_test[:, 2:]
X_new_A, X_new_B = X_test[:, :3], X_test[:, 3:]

history = model.fit((X_train_A, X_train_B), y_train, epochs=20,
                    validation_data=((X_valid_A, X_valid_B), y_valid))
mse_test = model.evaluate((X_test_A, X_test_B), y_test)
y_pred = model.predict((X_new_A, X_new_B))
```

### 10.2.5 서브클래싱 API로 동적 모델 만들기

사용할 충과 연결 방식을 먼저 정의해야 한다. 그다음 모델에 데이터를 주입하여 훈련이나 추론을 시작할 수 있다. 모델을 저장하거나 복사, 공유하기 쉽다.

```
class WideAndDeepModel(keras.Model):
  def __init(self, units=30, activation='relu', **kwargs):
    super().__init__(**kwargs)
    self.hidden1=keras.layers.Dense(units, activation = activation)
    self.hidden2=keras.layers.Dense(units, activation = activation)
    self.main_output = keras.kayers.Dense(1)
    self.aux_output = keras.layers.Dense(1)

  def call(self, inputs):
    input_A, input_B = inputs
    hidden1 = self.hidden1(input_B)
    hidden2 = self.hidden2(hidden1)
    concat = keras.layers.concatenate([input_A, hidden2])
    main_output = self.main_output(concat)
    aux_output = self.aux_output(hidden2)
    return main_output, aux_output

model = WideAndDeepModel()
```

Input 클래스의 객체를 만들 필요가 없다. call()메서드의 input 매개변수를 사용한다. 생성자에 있는 층 구성과 call()메서드에 있는 정방향 계산을 분리했다. 주된 차이점은 call()메서드 안에서 원하는 어떤 계산도 사용할수있다.
